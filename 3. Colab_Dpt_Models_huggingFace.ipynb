```python
#DPT (Dense Prediction Transformer) is being demonstrated with Depth Estimation Feature.
#In the code - Replace the model name appropriately.
#Dense Prediction Transformer (DPT) model trained on 1.4 million images for monocular depth estimation.
#It was introduced in the paper Vision Transformers for Dense Prediction by Ranftl et al. (2021) and first released in this repository.
#DPT uses the Vision Transformer (ViT) as backbone and adds a neck + head on top for monocular depth estimation.
#We can try these models in one go:
#dpt-large  - It is transformer based. Training data is mixed.  (https://huggingface.co/Intel/dpt-large)
#dpt-hybrid-midas - A CNN + Transformer making it hybrid. (https://huggingface.co/Intel/dpt-hybrid-midas)
#dpt-large-ade - Dense Prediction Transformer (DPT) model trained on ADE20k for semantic segmentation - (https://huggingface.co/Intel/dpt-large-ade)
#dpt-beit-large-512 - It is transformer based. (https://huggingface.co/Intel/dpt-beit-large-512)
#dpt-swinv2-large-384 - https://huggingface.co/Intel/dpt-swinv2-large-384

from transformers import pipeline
pipe = pipeline(task="depth-estimation", model="Intel/dpt-large")
result = pipe("http://images.cocodataset.org/val2017/000000039769.jpg")
result["depth"]
```
